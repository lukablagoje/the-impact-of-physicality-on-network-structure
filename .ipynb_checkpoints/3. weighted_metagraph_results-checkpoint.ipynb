## Identifying single neuron metagraph degree

```python
df_inverse = pd.read_csv('ME(R)_synapse_only_metagraph_edgelist_min_dist_threshold_weight.csv',index_col=[0])
df_inverse
```

```python
single_neuron_connectome_metagraph_dict = {}
single_neuron_bodyid = 1849136547
for radius in radius_list:
    #print(radius)
    radius_df = df_inverse[df_inverse['minimum_distance_threshold'] <= radius].copy()
    pair_list = []
    for row in radius_df.values:
        bodyid_pre = row[0]
        bodyid_post = row[1]
        if (bodyid_pre == single_neuron_bodyid) or (bodyid_post == single_neuron_bodyid):
            pair_list.append((bodyid_pre,bodyid_post))
    
    duplicate_list = []
    count = 0
    for pair in pair_list:
        if (pair[1],pair[0]) in pair_list:
             duplicate_list.append((pair[1],pair[0]))
        if pair in duplicate_list:
            count += 0
        else:
            count += 1
    if radius < 0.001:
        count = 0            
    single_neuron_connectome_metagraph_dict[radius]=count
    #connectome_metagraph_dict[radius] = 
```
