## 1. Creating a weighted metagraphs and a synaptic connectome (abstract network)
Below is the code to generate weighted threshold metagraphs - meaning that each pair of neurons will have a corresponding weight, which quantifies the minimum querying distance. In addition, the synaptic connectome is also obtained by using neuraprint library.

There are three possible criteria to do this:

1) No synapses - neurons which share a synapse will not be considered in the metagraph (as it can be considered trivial)

2) Only synapses - only neurons which share a synapse will be considered in the metagraph

3) Combined - all neurons will be considered

These criteria were developed to investigate how neurons that are connected synaptically compare to those that are not, in terms of physical distances and volume exclusion. 

```python
criteria_list =['no_synapses','only_synapses','combined']
criterium = 'no_synapses'
regioin_list = ['ME(R)']
for region in region_list:
    print('Region: ',region)
    radius_list = list(np.arange(0,1 + step ,0.01)) + [1, 2, 3, 4, 5, 10, 15, 20, 25, 30, 40, 45, 50, 60, 70, 80, 90, 100, 150,200]
    max_radius = max(radius_list)
    metagraph_radius_spherical_dictionary = {}
    connectome_radius_spherical_dictionary  = {}

    metagraph_links_spherical_dictionary = {}
    connectome_links_spherical_dictionary  = {}
    
    #Collecting all neurons that were queried in the previous notebook
    infile = open('querying_results/' + region +'_edge_dict_spherical_'+str(max_radius)+'.pkl','rb')
    metagraph_dictionary = pickle.load(infile)

    all_neurons = []
    for neuron_1 in metagraph_dictionary.keys():
        for neuron_2  in metagraph_dictionary[neuron_1]:
            all_neurons.append(neuron_1 )
            all_neurons.append(neuron_2)
    all_neurons = list(set(all_neurons))    
        
    # I create and store the connectome of all neurons used in the metagraph as CSV files 
    criteria = NC(bodyId=all_neurons)
    neuron_df,connectome_edgelist = fetch_adjacencies(criteria,criteria,properties=['size'])
    connectome_pair_list = connectome_edgelist[['bodyId_pre','bodyId_post']].values.tolist()
    
    # For each radius, a open labeled kd-tree querying results to find all neuron pairs that are collding
    for radius in radius_list:
        print('Radius',radius)
        infile = open('querying_results/' + region + '_edge_dict_spherical_'+str(radius)+'.pkl','rb')

        metagraph_dictionary = pickle.load(infile)
        metagraph_pair_list = []
        metagraph_neurons = []
        for neuron_1 in metagraph_dictionary.keys():
            for neuron_2 in metagraph_dictionary[neuron_1]:
                if criterium == 'no_synapses':
                    if not (([neuron_1,neuron_2] in connectome_pair_list ) or ([neuron_2,neuron_1] in connectome_pair_list )):
                        metagraph_pair_list.append([neuron_1,neuron_2])
                        metagraph_neurons.append(neuron_1)
                        metagraph_neurons.append(neuron_2)
                elif criterium =='only_synapses':
                    if [neuron_1,neuron_2] in connectome_pair_list or ([neuron_2,neuron_1] in connectome_pair_list:
                        metagraph_pair_list.append([neuron_1,neuron_2])
                        metagraph_neurons.append(neuron_1)
                        metagraph_neurons.append(neuron_2)                    
                elif criterium == 'combined':
                        metagraph_pair_list.append([neuron_1,neuron_2])
                        metagraph_neurons.append(neuron_1)
                        metagraph_neurons.append(neuron_2)    
                    
        metagraph_links_spherical_dictionary[radius] = metagraph_pair_list
        print('Number of neurons in the metagraph',len(set(metagraph_neurons))
              
    # Creating a minimum threshold weighted graph (when the neuron pair occurs for the first time)                                                          
    link_and_threshold_dict = {}
    for link in metagraph_links_spherical_dictionary[max_radius]:
        link_and_threshold_dict[(link[0],link[1])] = [max_radius]
                                                               
    # First, each distance is explored
    for minimum_distance in sorted(radius_list,reverse=True):
        for link in metagraph_links_spherical_dictionary[minimum_distance]:
            link_and_threshold_dict[(link[0],link[1])].append(minimum_distance)
                                                               
    # Afterward, for each link, the minimum (first occurence) distance is found
    link_and_threshold_list = []
    for link in link_and_threshold_dict.keys():
        min_threshold = min(link_and_threshold_dict[link])
        link_and_threshold_list.append([link[0],link[1],min_threshold])
    
    # I save the metagraph and connectome edgelist as CSV files
    weighted_metagraph_edgelist = pd.DataFrame(link_and_threshold_list,columns=['bodyId_1','bodyId_2','minimum_distance_threshold'])
    weighted_metagraph_edgelist.to_csv('metagraphs/' + criterium  + '_' + region + '_metagraph_edgelist_min_dist_threshold_weight.csv')
    
    # I create and store the connectome of all neurons used in the metagraph as CSV files                                                                                     
    connectome_edgelist = pd.DataFrame(connectome_links_spherical_dictionary[max_radius],columns=['bodyId_1','bodyId_2'])
    connectome_edgelist.to_csv('metagraphs/' + criterium  + '_' + region + '_connectome_edgelist.csv')
```

## 2. Results: Collision and distance properties of all neurons in the dataset
All the previous data processing is used to obtain relevant research results, which are displayed below:

### Number of colliding neurons below specific query distances

```python
# I load the previously obtained data
df_weighted_edgelist = pd.read_csv('metagraphs/' +'ME(R)_no_synapse_metagraph_edgelist_min_dist_threshold_weight.csv',index_col=[0])
# And I use the radii used in the metagraph creation process
step  = 0.01
end = 1.00
radius_list = list(np.arange(0,end  ,step))
radius_list_below = [np.round(i,2) for i in radius_list]
radius_list = radius_list_below  + [1, 2, 3, 4, 5, 10, 15, 20, 25, 30, 40, 45, 50, 60, 70, 80, 90, 100, 150,200]
```

On the left side, I show the number of neurons threshold radius from 0 (the resolution limit) to 1, while on the right side, I show the number of neurons for the threshold radius from 1 to 200 (redundant limit):

```python
plt.figure(figsize=(16,9))
plt.subplot(1,2,1)
plt.hist(df_weighted_edgelist[df_weighted_edgelist ['minimum_distance_threshold'] <1]['minimum_distance_threshold'],cumulative=True,bins= radius_list_below ,color='gray',alpha=0.8,edgecolor='black');
plt.xlabel('K-d tree query radius')
plt.ylabel('Count')
plt.subplot(1,2,2)
plt.hist(df_weighted_edgelist ['minimum_distance_threshold'],cumulative=True,bins=radius_list ,color='gray',alpha=0.8,edgecolor='black');
plt.xlabel('K-d tree query radius')
plt.ylabel('Count');
```

These plots show how many neurons would collide at a specific k-d tree query distances. For example, value 1 on the x-axis will show the number of neurons that would interesect at that query distance (or below that).

Even at the resolution limit, some neurons intersect (this due to the resolution limit). Additionally, there is a sharp rise of intersecting neurons are the threshold increases (from 1 to 10 k-d tree query radius).

### Metagraph edgelist and degree distribution
For each threshold value of distance, we can create a graph edgelist and analyze, like its degree distribution.

```python
# First I obtain the edgelist for all neurons with the k-d tree query distance less than 1:
metagraph_edgelist = df_weighted_edgelist[df_weighted_edgelist ['minimum_distance_threshold'] <1][['bodyId_1','bodyId_2']]
metagraph_edgelist
```

Then I can create a metagraph, where each edge corresponds to a collision:

```python
M = nx.Graph()
for edge in metagraph_edgelist.values:
    M.add_edge(edge[0],edge[1])
M_degree_dict = dict(M.degree())
plt.figure(figsize=(12,12))
nx.draw(M,alpha=0.5,node_size=25)
```

```python
plt.figure(figsize=(12,8))
plt.hist(M_degree_dict.values(),color='teal',alpha=0.8,edgecolor='black');
plt.xlabel("Metagraph degree")
plt.ylabel("Count");
```

There is one giant component, with many small separate components. Additionally, the degree distribution is sharp with most neurons having a low metagraph degree, while a minority of them have a high metagraph degree.

### Relationship between the connectome and the metagraph properties

```python
# I load the connectome edgelist
connectome_edgelist = pd.read_csv('metagraphs/' + 'ME(R)_connectome_edgelist.csv',index_col=[0])
connectome_edgelist 
```

```python
# And I use the networkx library to store it as a graph object
C= nx.Graph()
for edge in connectome_edgelist.values:
    C.add_edge(edge[0],edge[1])
C_degree_dict = dict(C.degree())
plt.figure(figsize=(12,12))
nx.draw(C,alpha=0.5,node_size=25,node_color ='red')
```

```python
plt.figure(figsize=(12,8))
plt.hist(C_degree_dict.values(),color='red',alpha=0.8,edgecolor='black');
plt.xlabel("Connectome degree")
plt.ylabel("Count");
```

```python
metagraph_degree_list = []
connectome_degree_list = []
for neuron in M_degree_dict.keys():
    if neuron in C_degree_dict:
        metagraph_degree_list.append(M_degree_dict[neuron])
        connectome_degree_list.append(C_degree_dict[neuron])
plt.figure(figsize=(12,8))
plt.scatter(metagraph_degree_list,connectome_degree_list,alpha=0.8,color='gray')
plt.xlabel("Metagraph degree")
plt.ylabel("Connectome degree")
```

There is a relationship between higher connectome and metagraph degree neurons.

## 3. Results for an individual neuron
The same process as above is repeated, but for an individual neuron. In the context of this research, there was a distiction between high and low metagraph degree neurons or rather highly and lowly confined neurons, which were analyzed in this manner:

```python
metagraph_no_synapse_edgelist= pd.read_csv('metagraphs/' + 'ME(R)_no_synapse_metagraph_edgelist_min_dist_threshold_weight.csv',index_col=[0])
metagraph_synapse_only_edgelist= pd.read_csv('metagraphs/' + 'ME(R)_only_synapse_metagraph_edgelist_min_dist_threshold_weight.csv',index_col=[0])
```

```python
single_neuron_synapse_only_metagraph_dict = {}
single_neuron_no_synapse_metagraph_dict= {}
# A particularly confined neuron was chosen
single_neuron_bodyid = 1849136547
for radius in radius_list:
    # I start the procedure for the metagraph with synapses
    radius_df =metagraph_synapse_only_edgelist[metagraph_synapse_only_edgelist['minimum_distance_threshold'] <= radius].copy()
    pair_list = []
    # I create a list of pairs for a specific radius value
    for row in radius_df.values:
        bodyid_pre = row[0]
        bodyid_post = row[1]
        if (bodyid_pre == single_neuron_bodyid) or (bodyid_post == single_neuron_bodyid):
            pair_list.append((bodyid_pre,bodyid_post))
    # Afterwards, I count the number of intersections, excluding duplicates
    duplicate_list = []
    count = 0
    for pair in pair_list:
        if (pair[1],pair[0]) in pair_list:
             duplicate_list.append((pair[1],pair[0]))
        if pair in duplicate_list:
            count += 0
        else:
            count += 1
    if radius < 0.001:
        count = 0            
    single_neuron_synapse_only_metagraph_dict [radius]=count
    
    
    # The same procedure is repeated for the metagraph without synapses
    radius_df =  metagraph_no_synapse_edgelist[metagraph_no_synapse_edgelist['minimum_distance_threshold'] <= radius].copy()
    pair_list = []
    # I create a list of pairs for a specific radius value
    for row in radius_df.values:
        bodyid_pre = row[0]
        bodyid_post = row[1]
        if (bodyid_pre == single_neuron_bodyid) or (bodyid_post == single_neuron_bodyid):
            pair_list.append((bodyid_pre,bodyid_post))
    # Afterwards, I count the number of intersections, excluding duplicates        
    duplicate_list = []
    count = 0
    for pair in pair_list:
        if (pair[1],pair[0]) in pair_list:
             duplicate_list.append((pair[1],pair[0]))
        if pair in duplicate_list:
            count += 0
        else:
            count += 1

    if radius < 0.001:
        count = 0            
    single_neuron_no_synapse_metagraph_dict[radius]=count
```

```python
plt.figure(figsize=(12,8))
plt.plot(single_neuron_synapse_only_metagraph_dict .keys(),single_neuron_synapse_only_metagraph_dict.values(),label='Synapse only metagraph')
plt.plot(single_neuron_no_synapse_metagraph_dict.keys(),single_neuron_no_synapse_metagraph_dict.values(),label='No synapse metagraph')
plt.ylabel('Number of links')
plt.xlabel('K-d tree query radius')
plt.title("High metagraph degree neuron " +str(single_neuron_bodyid ))
plt.legend();
```

```python
plt.figure(figsize=(12,8))
plt.plot(single_neuron_synapse_only_metagraph_dict .keys(),single_neuron_synapse_only_metagraph_dict.values(),label='Synapse only metagraph')
plt.plot(single_neuron_no_synapse_metagraph_dict.keys(),single_neuron_no_synapse_metagraph_dict.values(),label='No synapse metagraph')
plt.ylabel('Number of links')
plt.xlabel('K-d tree query radius')
plt.title("High metagraph degree neuron " +str(single_neuron_bodyid ))
plt.legend()
plt.yscale('log')
plt.xscale('log')
```

As expected, metagraph with synaptically connected neurons is quicker to increase for this single neuron, but tops off (as the maximum number is attained). This is not true for the metagraph degree, which just keeps steadily increasing with the k-d tree query distance.
